```python

TAG_CHARS = list(range(0xE0000, 0xE007F + 1))  # Unicode tag characters
VARIATION_SELECTORS = [*range(0xFE00, 0xFE0F + 1), *range(0xE0100, 0xE01EF + 1)]  # Variation selectors
BIDI_CHARS = [*range(0x200E, 0x200F+1),*range(0x202A, 0x202E + 1), *range(0x2066, 0x2069 + 1)]  # Bidi control characters
DEL_CHAR = [0x0008]  # Deletion character
INVIS_CHAR = [0x200b, 0x200c, 0x200d, 0xFEFF, 0x2060] #zero-widths


ALL_INVISIBLE = list(TAG_CHARS + VARIATION_SELECTORS + BIDI_CHARS + DEL_CHAR + INVIS_CHAR)
codepoints = [c for c in ALL_INVISIBLE]
print(codepoints)
```

```python
text = ''',Temüjin's️ reputation󠀵 also︊ beca‎me wide󠄵󠄠spread ‭afte󠇎r h󠀘is escap‌e from th﻿‮e Tayich﻿i'ud󠇘‏󠄝.‎󠄿,'''
text_cp= ([ord(c) for c in text])

count = 0
for c in text_cp:
    if c in codepoints:
        count += 1
        print(c)
print("Count", count)
print("Text code points" ,text_cp)
```
