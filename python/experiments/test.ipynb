```python
import pickle

# Open the file in read-binary mode
with open("adversarial-examples.pkl", "rb") as file:
    data = pickle.load(file)

```

```python
# Reorderings
reorderings = []
for perturbation_level in data["perspective_reorderings"]:
    if perturbation_level != "0":
        for id in data["perspective_reorderings"][perturbation_level]:
            adversarial_example = data["perspective_reorderings"][perturbation_level][
                id
            ]["adv_example"]
            input = data["perspective_invisibles"][perturbation_level][id]["input"]
            reorderings.append({"input":input, "adv_example": adversarial_example})

p = _pipeline.Pipeline()
p.add_bidi_sanitizer()
p.add_invisible_sanitizer()
p.add_tag_sanitizer()
p.add_variation_selector_sanitizer()

print(len(reorderings))
for reordering_sample in reorderings:
    input = reordering_sample["input"]
    adversarial_example = reordering_sample["adv_example"]
    assert(input != adversarial_example)
    assert input == p.sanitize(adversarial_example)
```

```python
#Invisibles
invisibles = [] 
for perturbation_level in data["perspective_invisibles"]:
    if perturbation_level == "5":
        for id in data["perspective_invisibles"][perturbation_level]:
            adversarial_example = data["perspective_invisibles"][perturbation_level][id]["adv_example"]
            input = data["perspective_invisibles"][perturbation_level][id]["input"]
            invisibles.append({"input": input, "adv_example": adversarial_example})

p = _pipeline.Pipeline()
p.add_bidi_sanitizer()
p.add_invisible_sanitizer()
p.add_tag_sanitizer()
p.add_variation_selector_sanitizer()

print(len(reorderings))
for reordering_sample in reorderings:
    input = reordering_sample["input"]
    adversarial_example = reordering_sample["adv_example"]
    assert input != adversarial_example
    assert input == p.sanitize(adversarial_example)
```

```python
# Deletions
deletions = []
for perturbation_level in data["perspective_deletions"]:
    if perturbation_level != "0":
        for id in data["perspective_deletions"][perturbation_level]:
            adversarial_example = data["perspective_deletions"][perturbation_level][id][
                "adv_example"
            ]
            input = data["perspective_deletions"][perturbation_level][id]["input"]
            deletions.append({"input": input, "adv_example": adversarial_example})

p = _pipeline.Pipeline()
p.add_bidi_sanitizer()
p.add_invisible_sanitizer()
p.add_tag_sanitizer()
p.add_variation_selector_sanitizer()
p.add_deletion_sanitizer()

print(len(reorderings))
for deletion_sample in deletions:
    input = deletion_sample["input"]
    adversarial_example = deletion_sample["adv_example"]
    assert input != adversarial_example
    assert input == p.sanitize(adversarial_example)
```
